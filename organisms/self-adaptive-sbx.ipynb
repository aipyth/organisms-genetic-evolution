```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def eta(u, beta_a):
  et = - (1 + (np.log(2 * (1 - u))) / (np.log(beta_a)))
  return np.min(50, np.max(0, et))
```

```python
c1 = np.random.random()
c2 = np.random.random()
p22 = np.random.random()
alpha = 1.5
beta_a = np.abs(1 + (2 * alpha * (c2 - p22)) / (c2 - c1))

print(beta_a)

u = np.linspace(0, 1, 50)[1:-1]

plt.plot(u, eta(u, beta_a))
```
